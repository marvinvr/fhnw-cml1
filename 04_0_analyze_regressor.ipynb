# Analyze regressor

## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
from helpers.paths import Paths
from joblib import load
```

## Model Import

```python
best_model = load(Paths.REGRESSOR_MODEL_DATA('gradientboostingregressor'))
best_model
```

## Analyze feature importance

```python
importances = best_model.feature_importances_
features = best_model.feature_names_in_
feature_importances = pd.DataFrame({"Features": features, "Importance": importances})
feature_importances = feature_importances.sort_values(by = "Importance", ascending = False)

feature_importances.head(20).plot.bar(x='Features', y='Importance')
plt.title('Wichtigkeit der Features (Variable Importance)')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.legend().remove()
plt.tight_layout()
plt.show()
```

Im Obigen Plot sind man die 20 wichtigsten Features aufgelistet. Die Features sind nach ihrer Wichtigkeit sortiert.
Die Wichtigkeit gibt an, wie stark ein Feature dazu beiträgt, die Vorhersage des Modells zu verbessern.
